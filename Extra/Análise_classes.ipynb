```python
import pandas as pd
from sqlalchemy import create_engine
from IPython.display import display

# Conectar ao banco de dados PostgreSQL
engine = create_engine('postgresql://joaonolasco:32313167@localhost:5432/meu_banco')

# Executar a consulta SQL para filtrar apenas os com baixa cobertura a partir da camada gold
query_analise_produtos_classe_baixa_gold = """
SELECT 
    classe,
    total_bairros_atendidos,
    CASE 
        WHEN total_bairros_atendidos < 10 THEN 'Baixa cobertura'
        WHEN total_bairros_atendidos BETWEEN 10 AND 30 THEN 'Cobertura média'
        ELSE 'Alta cobertura'
    END AS cobertura_bairro
FROM gold.produtos_classe_completo
WHERE total_bairros_atendidos < 10
ORDER BY total_bairros_atendidos DESC;
"""

# Carregar os resultados da consulta no DataFrame
df_analise_produtos_classe_baixa_gold = pd.read_sql(query_analise_produtos_classe_baixa_gold, engine)

# Exibir o resultado filtrado (apenas baixa cobertura)
print("Análise de Produtos por Classe com Baixa Cobertura (Schema Gold)")
display(df_analise_produtos_classe_baixa_gold)
```
